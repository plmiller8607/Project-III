# Data from SQL Queries



```python
import matplotlib.pyplot as plt
```

```python

#Overall Graduation Rate: What is the overall graduation rate across all schools for Maricopa County in 2023?
# Top Schools: Which schools had the highest graduation rates?
# Lowest Schools: Which schools had the lowest graduation rates?

schools = [
    "Overall Graduation Rate",
    "Shadow Ridge High School",
    "Mountain Ridge High School",
    "Willow Canyon High School",
    "Sandra Day O'Connor High School",
    "Liberty High School",
    "EdOptions HS Learning Center",
    "Pathways in Education",
    "Ombudsman - Charter Metro",
    "Ombudsman - Charter Valencia",
    "Arizona Preparatory Academy"
]

grad_rates = [
    74.8,
    98,
    98,
    97,
    97,
    97,
    14,
    14,
    13,
    12,
    9
]

# Create bar chart
plt.figure(figsize=(13, 6))
bars = plt.barh(schools, grad_rates, color='skyblue')
plt.xlabel('Graduation Rate (%)')
plt.title('High Schools with Highest & Lowest Graduating Rates')

# Invert y-axis to have the highest rate on top
plt.gca().invert_yaxis() 

# Add percentage labels
for bar in bars:
    width = bar.get_width()
    label_y = bar.get_y() + bar.get_height() / 2
    plt.text(width + 0.5, label_y, f'{width:.1f}%', va='center')

# Save the plot as a file
plt.savefig('highest_lowest_chart.png')

plt.show()
```

```python
# Graduation Rate by Subgroup: How does the graduation rate vary by subgroups?


subgroups = [
    "Overall Graduation Rate",
    "Female",
    "Male",
    "American Indian or Alaska Native",
    "Black/African American",
    "Hispanic or Latino",
    "Multiple Races",
    "White",
    "English Learners",
    "Homeless",
    "Military",
    "Special Education"
]

graduation_rates = [
    74.8,
    76.7,
    71.5,
    60.8,
    60.9,
    70.1,
    27.6,
    82.0,
    67.2,
    39.4,
    35.5,
    68.3
]
# Color mapping
colors = [
    # Overall Graduation Rate,
    'skyblue',
    # Female, Male
    'red', 'red', 
    # American Indian or Alaska Native, Black/African American, Hispanic or Latino, Multiple Races, White
    'blue', 'blue', 'blue', 'blue', 'blue', 
    # English Learners, Homeless, Military, Special Education
    'green', 'green', 'green', 'green'  
]

# Create bar chart
plt.figure(figsize=(13, 9))
bars = plt.barh(subgroups, graduation_rates, color=colors)
plt.xlabel('Graduation Rate (%)')
plt.title('Graduation Rates by Subgroup')
plt.gca().invert_yaxis()  # Highest rate on top

# Add percentage labels
for bar in bars:
    width = bar.get_width()
    label_y = bar.get_y() + bar.get_height() / 2
    plt.text(width + 0.5, label_y, f'{width:.1f}%', va='center')

plt.savefig('rates_by_subgroups_chart.png')
plt.show()
```
